## 1.

Write a function that returns a circuit, that only uses $H$ and $T$ gates, that calculates the state:

$$|\psi \rangle = \frac{1}{\sqrt{2}} ( |0\rangle - i |1\rangle )$$


```
def circuit_1():
    from qiskit import QuantumCircuit
    qc = QuantumCircuit(1)
    # your code here

    # Hadamard gate to create a superposition
    qc.h(0)

    # Apply the T gate six times
    for _ in range(6):
      qc.t(0)

    return qc
```

## 3.

Write a function that returns a circuit that calculates the classic OR gate when aplied to the states $|00\rangle$, $|01\rangle$, $|10\rangle$ and $|11\rangle$. The circuits must have 3 qubits: qubit 0 and 1 for the inputs and qubit 2 for the output. The final state of qubits 0 and 1 must be same as the input. The circuit can only use X and CCX gates.

```
def circuit_3():
    from qiskit import QuantumCircuit
    qc = QuantumCircuit(3)
    # your code here

    # Apply NOT to both inputs
    qc.x(0)
    qc.x(1)

    # Compute (NOT A) AND (NOT B) using a Toffoli gate
    qc.ccx(0, 1, 2)

    # Apply NOT to the result to get A OR B
    qc.x(2)

    # Uncompute the NOT gates on the inputs to restore them
    qc.x(1)
    qc.x(0)

    return qc
```

## 6.

Write a function that returns a circuit with the following unitary matrix:

$$
\text{U} = \begin{bmatrix}
\tfrac{1}{2} & \tfrac{1}{2} & -\tfrac{1}{2} & -\tfrac{1}{2}  \\
\tfrac{1}{2} & -\tfrac{1}{2} & -\tfrac{1}{2} & \tfrac{1}{2}  \\
\tfrac{1}{2} & \tfrac{1}{2} & \tfrac{1}{2} & \tfrac{1}{2}  \\
-\tfrac{1}{2} & \tfrac{1}{2} & -\tfrac{1}{2} & \tfrac{1}{2}  \\
\end{bmatrix}
$$

The circuit can only use $H$, $X$ and $CNOT$ gates.

```
def circuit_6():
    from qiskit import QuantumCircuit
    qc = QuantumCircuit(2)
    # your code here
    # This circuit implements the decomposition:
    # U = CZ(1,0) * X(1) * H(0) * H(1)
    #
    # We build CZ(1,0) from H and CX (CNOT):
    # CZ(1,0) = H(0) * CX(1,0) * H(0)
    #
    # So the full gate sequence (read from right to left in the math) is:
    # U = [H(0) * CX(1,0) * H(0)] * X(1) * H(1) * H(0)
    #
    # In Qiskit, this translates to (from top to bottom):

    # 1. H on both qubits (H(1) * H(0))
    qc.h(0)
    qc.h(1)

    # 2. X on qubit 1 (X(1))
    qc.x(1)

    # 3. CZ(1,0) gate: H(0) * CX(1,0) * H(0)
    qc.h(0)
    qc.cx(1, 0) # Control=q1, Target=q0
    qc.h(0)

    return qc
```

## 7.

Write a function that returns a circuit that calculates the following state:

$$|\psi \rangle = \frac{1}{\sqrt{3}} ( |001\rangle + |010\rangle + |100\rangle  )$$

The circuit can use the $CNOT$, $RY$, $H$ and $T$ gates.


```
def circuit_7():
    from qiskit import QuantumCircuit
    qc = QuantumCircuit(3)
    ## Your code here
    # This state is a W-state, |W_3>, |psi> = 1/sqrt(3) * (|001> + |010> + |100>)
    # Qiskit is little-endian, so this is:
    # 1/sqrt(3) * (|q2=0,q1=0,q0=1> + |q2=0,q1=1,q0=0> + |q2=1,q1=0,q0=0>)

    # function for X gate using H and T
    def x_gate(circuit, qubit):
        """Applies an X gate using H and T gates."""
        circuit.h(qubit)
        circuit.t(qubit)
        circuit.t(qubit)
        circuit.t(qubit)
        circuit.t(qubit)
        circuit.h(qubit)

    # unction for CCX gate (Toffoli)
    def ccx_gate(circuit, c1, c2, t):
        """Applies a CCX gate decomposed into H, T, T-dagger, and CX."""

        # T-dagger gate (T^3)
        def t_dagger(circuit, qubit):
            circuit.t(qubit)
            circuit.t(qubit)
            circuit.t(qubit)

        circuit.h(t)
        circuit.cx(c2, t)
        t_dagger(circuit, t)
        circuit.cx(c1, t)
        circuit.t(t)
        circuit.cx(c2, t)
        t_dagger(circuit, t)
        circuit.cx(c1, t)
        circuit.t(t)
        circuit.t(c2)
        circuit.h(t)
        circuit.cx(c1, c2)
        circuit.t(c1)
        t_dagger(circuit, c2)
        circuit.cx(c1, c2)

    # Apply a Ry rotation to qubit 0 to get the |001> component.

    theta_1 = 2 * np.arcsin(np.sqrt(1/3))
    qc.ry(theta_1, 0)
    # State is now: sqrt(2/3)|000> + sqrt(1/3)|001>

    # Apply a controlled-Ry rotation on qubit 1, controlled by qubit 0.


    theta_2 = np.pi / 2

    # Apply X(0)
    x_gate(qc, 0)

    # Apply C_1-RY(pi/2, 0, 1)
    qc.ry(theta_2 / 2, 1) # ry(pi/4, 1)
    qc.cx(0, 1)
    qc.ry(-theta_2 / 2, 1) # ry(-pi/4, 1)
    qc.cx(0, 1)

    # Apply X(0)
    x_gate(qc, 0)
    # State is now: sqrt(1/3)|000> + sqrt(1/3)|010> + sqrt(1/3)|001>

    # Apply a doubly-controlled-NOT on qubit 2.
    #    We want to apply X(2) if q1 is 0 AND q0 is 0.
    #    This is C_0-C_0-X(c1=q1, c2=q0, t=q2)
    #    This is: X(c1) X(c2) CCX(c1, c2, t) X(c2) X(c1)

    # Apply X(0)
    x_gate(qc, 0)
    # Apply X(1)
    x_gate(qc, 1)

    # Apply CCX(1, 0, 2)
    ccx_gate(qc, 1, 0, 2)

    # Apply X(1)
    x_gate(qc, 1)
    # Apply X(0)
    x_gate(qc, 0)

    return qc
```
